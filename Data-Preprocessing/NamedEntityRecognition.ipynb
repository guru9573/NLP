# Named Entity Recognition

* It is used to extract information from unstructured text.
* It is used to classify the entities present in the text into categories such as person, organization, event places etc., 
* This will give detailed knowledge about the text and the relationship between entities

```python
import nltk

from nltk.tokenize import word_tokenize
from nltk import pos_tag, ne_chunk
```

```python
nltk.download('punkt')
nltk.download('maxent_ne_chunker')
nltk.download('words')
nltk.download('averaged_perceptron_tagger')
```

```python
def ner(text):
  tokens = word_tokenize(text)
  pos = pos_tag(tokens)

  return ne_chunk(pos)
```

```python
input_text= "My name is Sai Shiva"

print(ner(input_text))
```

```python
import spacy 
nlp = spacy.load("en_core_web_sm")

doc = nlp("NASA awarded Elon Musk’s SpaceX a $2.9 billion contract to build the lunar lander.")
for token in doc:
    print(token.text, token.ent_iob_, token.ent_type_)
```
