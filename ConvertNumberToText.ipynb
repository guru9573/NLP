```python
import inflect
```

```python
engine = inflect.engine()
```

```python
def convert_num_to_str(text):
  temp_str = text.split()

  new_str = []

  for word in temp_str:
    if word.isdigit():
      num_text = engine.number_to_words(word)
      new_str.append(num_text)
    else:
      new_str.append(word)

  return ' '.join(new_str)
```

```python
convert_num_to_str('I have 5352 rupees')
```
